```python
from groq import Groq
from google.colab import userdata

client = Groq(api_key=userdata.get("GROQ_API_KEY"))

user_text =(input("enter your text: "))
target_lvl=(input("\n enter the target reading level(i.e: child, teenager, or adult): ")).lower()

system_prompt = None

if target_lvl == "child":
  system_prompt = "you're a text rewriter who rewrites a text given by user in a child's language"
elif target_lvl == "teenager":
  system_prompt = "you're a text rewriter who rewrites a text given by user in a teenager's language"
elif target_lvl == "adult":
  system_prompt = "you're a text rewriter who rewrites a text given by user in an adult's language"
else:
  print("invalid input")

if system_prompt is not None:
  chat= client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role":"system", "content" : system_prompt},
         {"role":"user", "content": user_text}
    ]
)
  print(chat.choices[0].message.content)
```
